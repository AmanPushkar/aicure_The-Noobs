```python
#droppping all columns with categorical variables for numerical data analysis
input = train_df.drop(['uuid','condition','datasetId','HR'],axis=1)
```

performing numerical data analysis



```python
# finding null values(if any)

def data_plots(dataframe, target_output, columns,figsize=(30,60)):

    for i, column in enumerate(columns):
        plt.figure(figsize=figsize)
        plt.subplot(3,12, i + 1)
        plt.scatter(dataframe[column], target_output, color='green', marker='.')
        plt.xlabel(column)
        plt.ylabel('HR')
        plt.show()
```

```python
def correlation_determinance(df, target):
    cols = df.columns.tolist()

    for i in range(len(cols)):
        correlation = df[cols[i]].corr(target)
        print(f'{cols[i]} vs {target.name}: {correlation}')
```

```python
input_features = input.columns
```

```python
data_plots(input,train_df['HR'],input_features)
```

```python
correlation_determinance(input,train_df['HR'])
```

```python
#considering the subplots we can infer that the given dataset will not perform well for
#a tree based model or any gaussian based models and the data is linear to use deep learning algorithms but
#the data is not too linear for normal linear models so our best approach will be the use of polynomial regression
#given the covariance relations with the HR all but some features seem to not contribute to the output feature
```

```python
le = LabelEncoder()
condition = le.fit_transform(train_df['condition'])
condition = pd.DataFrame(condition)
condition['condition'] = condition
condition = condition.drop(0 , axis=1)
```

```python
# preparing the data for training and testing purposes
```

```python
input = pd.concat([input,condition],axis=1)
input = input.drop('SDRR_RMSSD',axis=1)
```

```python
input.head()
```

```python
#splitting the data
x_train,x_test,y_train,y_test = train_test_split(input,train_df['HR'], test_size = 0.2 , random_state=42)
```

```python
polynomial_features = PolynomialFeatures(degree=3, include_bias=False)
linear_regression = LinearRegression()

poly_regression = Pipeline([
        ('polynomial_features', polynomial_features),
        ('linear_regression', linear_regression)
    ])
```

```python
poly_regression.fit(x_train,y_train)
```

```python
poly_regression.score(x_test,y_test)
```

```python
y_pred = poly_regression.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2): {r2}')
```

```python
# the polynomial model is working fine with good r2 and mse errors
```

```python
polynomial_features = PolynomialFeatures(degree=3, include_bias=False)
linear_regression = LinearRegression()

poly_final_model = Pipeline([
        ('polynomial_features', polynomial_features),
        ('linear_regression', linear_regression)
    ])
```
